## Task 1: Prediction using Supervised ML
Predict the percentage score of a student based on number of study hours

### First importing all the required libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading data from remote link 

```python
url="http://bit.ly/w-data"
s_data=pd.read_csv(url)
print("data imported successfully")

s_data.head()
```

### Plotting the distribution of scores

```python
s_data.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()
```

From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.

### Making Predictions
Here we predict the scores from our algorithm

```python
X = s_data.iloc[:, :-1].values  
Y = s_data.iloc[:, 1].values  
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                            test_size=0.2, random_state=0) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg=LinearRegression()
reg=reg.fit(X_train, Y_train)

print("Training Complete")
```

### Plotting the Regression line

```python
line = reg.coef_*X+reg.intercept_

plt.scatter(X, Y)
plt.plot(X, line, c='000000');
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score') 
plt.show()
```

### Actual vs Predicted data 

```python
print(X_test) 
Y_pred = reg.predict(X_test)
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})  
df
```

### Predicting the score from the given data (number of hours 9.25)

```python
hours = [[9.25]]
own_pred = reg.predict(hours)
print("No of Hours = {}".format(hours[0][0]))
print("Predicted Score = {}".format(own_pred[0]))
```

### Evaluating the model
In the final step performance of the algorithm is evaluated. We choose Mean Absolute Error, Mean Squared error and Root Mean Squared Error as a metric

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(Y_test,Y_pred)) 
print('\n')
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
print('\n')
print('Root Mean Squared Error:',  np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
